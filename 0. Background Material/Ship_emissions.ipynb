the emissions are taken at the time of satellite overpasses and then the satellite data is taken at the emissions' locations

for a given day, we first load the ship locations of the entire day and the day before 

```python
print("working on day:", date.strftime("%Y-%m-%d"))

emissions_df = pd.read_csv(emissions_path/f"{date.year:04d}"/f"{date.month:02d}_{date.day:02d}")
                           
day_before = date - timedelta(days=1)
try:
    emissions_before = pd.read_csv(emissions_path/f"{day_before.year:04d}"/f"{day_before.month:02d}_{day_before.day:02d}")
    
    # we add an arbitrary number to the particle column, which IDs individual ships, to distinguish the day and the day before
    emissions_before.particle=emissions_before.particle+10000 
    
    emissions_df=pd.concat([emissions_before, emissions_df])

except FileNotFoundError: 
    print("location file not found for", date.strftime("%Y-%m-%d"))

emissions_df=emissions_df.set_index(pd.to_datetime(emissions_df['hour']))
emissions_df=emissions_df.drop(columns=['hour']).sort_index()
```

for each of the 24h preceding the overpass, we get the emissions from that time, load the file that contains the advected emissions and finally replace the location of emission with the simulated location at the time of interest

```python
advection_path_eo1 = pathlib.Path("/gws/nopw/j04/eo_shared_data_vol1/satellite/modistracks/hysplit_adv_20m_non_isob")
advection_path_eo2 = pathlib.Path("/gws/nopw/j04/eo_shared_data_vol2/scratch/pete_nut/hysplit_adv_20m_non_isob")
```

```python
for i, t in enumerate(np.unique(emissions_advected.index)):
    print(i+1, '/', len(np.unique(emissions_advected.index)))

    start_time = pd.to_datetime(t).to_pydatetime()

    # Skip this loop if time_of_overpass is less than the emissions start time
    if (time_of_overpass - start_time).total_seconds() < 0: 
        continue
    
    try:
        # if start_time.year in (2017, 2018, 2019):
        #     filloc = advection_path_eo2/f"{start_time.year:04d}_{start_time.month:02d}_{start_time.day:02d}_{start_time.hour:02d}:{start_time.minute:02d}"
        # else: 
        filloc = advection_path_eo1/f"{start_time.year:04d}_{start_time.month:02d}_{start_time.day:02d}_{start_time.hour:02d}:{start_time.minute:02d}"
        # this csv file has the locations where emissions from year, month, day, 
        # start_time end up at any given moment within 24h in 5min steps
        trajectories_df = pd.read_csv(filloc)
        
        trajectories_df.loc[:,"jday"] = pd.to_datetime(trajectories_df.jday)
        
        trajectories_df = trajectories_df[trajectories_df.jday==time_of_overpass]
        
        if len(trajectories_df) != len(emissions_advected[emissions_advected.index==t]):
            for i in range(1, len(emissions_advected[emissions_advected.index==t]) + 1):
                if not np.isin(i, np.unique(trajectories_df.particle)):
                    print('missing particle', i)
                    new_row = {'particle':i, 'jday':time_of_overpass, 'lat':np.nan, 'lon':np.nan , 'alt':np.nan, 'timestep':np.nan }
                    #append row to the dataframe
                    trajectories_df = pd.concat([trajectories_df, pd.DataFrame([new_row])], ignore_index=True)
                    trajectories_df = trajectories_df.sort_values('particle') 
                else: None
        else: None
        emissions_advected.loc[emissions_advected.index==t, "latitude"] = trajectories_df.lat.values
        emissions_advected.loc[emissions_advected.index==t, "longitude"] = trajectories_df.lon.values

    except FileNotFoundError:
        emissions_advected.latitude[emissions_advected.index==t] = np.nan
        emissions_advected.longitude[emissions_advected.index==t] = np.nan
        print('no hysplit data for {}'.format(start_time))
```

# Load SEVIRI and overlay ship tracks

SEVIRI (spinning enhanced visible and infra-red imager) is an instrument located in a geostationary orbit aboard the Meteosat second generation (MSG) weather satellites. It takes an image of the Earth every 15 minutes, with 3km resolution, in a range of wavelengths including visible and IR light. We can use these different measurements to make retrievals of cloud properties

We only have these retrievals for 2019, so lets focus on that year

```python
seviri_path = pathlib.Path("/gws/nopw/j04/eo_shared_data_vol1/satellite/seviri-orac/Data")
```

We only have retrieved cloud properties for 2019, so lets start with those for now

```python
seviri_overpass_time = datetime(2019, 1, 3, 8, 30)
```

We have 4 different file types that contain different variables:

- ESACCI-L2_CLOUD-CLD_PRODUCTS: Retrieved cloud properties, including height, temperature, optical thickness, effective radius and liquid water path
- ESACCI-TOA: Derived broadband fluxes for the top- and bottom-of-atmosphere, including both cloudy and clear sky fluxes. These variables can be used to show the radiative effect of the observed clouds on the atmosphere
- SEVIRI_ML: First guess cloud properties predicted using an ML model. These probably won't be of any use
- SECONDARY: Input data and processing variables from the cloud retrieval. This mostly won't be of any use, but the input data can be used to show images from seviri

```python
seviri_files = list((seviri_path/f"{seviri_overpass_time.year:04d}"/f"{seviri_overpass_time.month:02d}"/f"{seviri_overpass_time.day:02d}"/f"{seviri_overpass_time.hour:02d}").glob(f"*{seviri_overpass_time.strftime('%Y%m%d%H%M')}*.nc"))
seviri_files
```

```python
secondary_ds = xr.open_dataset('/gws/nopw/j04/eo_shared_data_vol1/satellite/seviri-orac/Data/2019/01/03/08/SECONDARY-20190103083000-ESACCI-L2_CLOUD-CLD_PRODUCTS-SEVIRI-MSG4-fv3.0.nc')
secondary_ds
```

First let's plot a "False colour" image from SEVIRI. We use channels 3, 2 and 1 and the red, green and blue channels respectively. The wavelengths of these channels are actually 1.6Î¼m, 0.81Î¼m and 0.64Î¼m, so we are using a combination of near-IR and red light. Ice clouds appear cyan because they reflect less at longer wavelengths

SEVIRI scans bottom to top and right to left, which is the opposite of a conventional camera. As a result we need to flip the image in both axes to plot it the correct way up

```python
gamma = 0.5
RGB = np.maximum(np.minimum(np.stack(
    [
        secondary_ds.reflectance_in_channel_no_3, 
        secondary_ds.reflectance_in_channel_no_2, 
        secondary_ds.reflectance_in_channel_no_1,
    ],
    -1
), 1), 0)**gamma

geo_p = ccrs.Geostationary(central_longitude=0.0, satellite_height=35785831, false_easting=0, false_northing=0, globe=None)
geo_extent = ( geo_p.x_limits[0], geo_p.x_limits[1], geo_p.y_limits[0], geo_p.y_limits[1] )

fig = plt.figure()
ax = plt.axes(projection=geo_p)

ax.imshow(RGB[::-1,::-1][51:3660, 51:3667], extent=geo_extent)
ax.coastlines(alpha=0.5)
title = plt.title("SEVIRI False RGB visible image")
```

As the time is during the morning, part of the image is in darkness.

We can also look at the Earth in the longwave IR. Channel 9 has a wavelength of 10.8Î¼m, which is in the "longwave window" where there is little absorption by the atmosphere. As this light is emitted, rather than reflected, it will appear the same during both day and night. High clouds appear darker due to their lower temperature emissions. Low clouds are hard to see as they have a similar temperature to the surface

```python
fig = plt.figure()
ax = plt.axes(projection=geo_p)

img = ax.imshow(secondary_ds.brightness_temperature_in_channel_no_9[::-1,::-1][51:3660, 51:3667], extent=geo_extent, vmin=180, vmax=320, cmap="inferno")
ax.coastlines(alpha=0.5)
plt.colorbar(img, label="10.8Î¼m brightness temperature [K]")
title = plt.title("SEVIRI channel 9 IR image")
```

Reflectances and brightness temperatures only give us a qualitative view of the clouds. Let's look at some of the retrieved cloud properties to get a more quantitative view

```python
cloud_ds = xr.open_dataset('/gws/nopw/j04/eo_shared_data_vol1/satellite/seviri-orac/Data/2019/01/03/08/20190103083000-ESACCI-L2_CLOUD-CLD_PRODUCTS-SEVIRI-MSG4-fv3.0.nc', decode_times=False)
cloud_ds
```

Here we plot a variety of the cloud properties available

Cloud top height and cloud top temperature both show similar properties, albeit inverted, as higher clouds tend to also be older. Note that in both these cases we use the corrected versions of these properties. The light reflected/emitted by the cloud top does not come from the very top of the clouds, but a layer within that can be quite deep for thin clouds with low reflectivity/emissivity. As a result, the retrieved CTH/CTT does not represent the very top of the clouds, unless corrected based off of their optical properties. Both retrieved and corrected cloud height properties are available

Cloud optical thickness (COT) represents the transmissivity of the cloud, with a larger COT meaning less light is transmitted.

Cloud particle effective radius (r_e) represents the scattering cross-section of the cloud particles. Smaller particles tend to scatter more light, so a given cloud will tend to have a larger COT if it has smaller r_e.

Note that while the optical properties are retrieved for both day- and nighttime, they are less accurate at night because no visible information is available. This is particularly important for r_e, as it has a much larger effect on reflectivity than emissivity

```python
fig, axes = plt.subplots(2, 2, figsize=(12,9.6), subplot_kw={"projection":geo_p})

img = axes[0,0].imshow(cloud_ds.cth_corrected[::-1,::-1][51:3660, 51:3667], extent=geo_extent, vmin=0, vmax=20, cmap="Greens")
axes[0,0].coastlines(alpha=0.5)
plt.colorbar(img, label="Cloud top height [km]", ax=axes[0,0])
title = axes[0,0].set_title("Corrected cloud top height")

img = axes[0,1].imshow(cloud_ds.ctt_corrected[::-1,::-1][51:3660, 51:3667], extent=geo_extent, vmin=180, vmax=320, cmap="Oranges")
axes[0,1].coastlines(alpha=0.5)
plt.colorbar(img, label="Cloud top temperature [K]", ax=axes[0,1])
title = axes[0,1].set_title("Corrected cloud top temperature")

img = axes[1,0].imshow(cloud_ds.cot[::-1,::-1][51:3660, 51:3667], extent=geo_extent, vmin=0, vmax=50, cmap="Blues")
axes[1,0].coastlines(alpha=0.5)
plt.colorbar(img, label="Cloud optical thickness", ax=axes[1,0])
title = axes[1,0].set_title("Cloud optical thickness")

img = axes[1,1].imshow(cloud_ds.cer[::-1,::-1][51:3660, 51:3667], extent=geo_extent, vmin=0, vmax=50, cmap="Purples")
axes[1,1].coastlines(alpha=0.5)
plt.colorbar(img, label="Particle effective radius [Î¼m]", ax=axes[1,1])
title = axes[1,1].set_title("Cloud particle effective radius")
```

Finally, we can look at the broadband fluxes which show how the Earth and its atmosphere are being heated at different levels

```python
flux_ds = xr.open_dataset('/gws/nopw/j04/eo_shared_data_vol1/satellite/seviri-orac/Data/2019/01/03/08/20190103083000-ESACCI-TOA-SEVIRI-MSG4-fv3.0.nc', decode_times=False)
flux_ds
```

The top-of-atmoshere (ToA) flux balance is calculated as the incoming solar, shortwave (SW), radiation minus the outgoing SW and LW radiation. During the daytime this tends to be positive due to the incoming solar radiation, but is negative at night.

```python
toa_flux = flux_ds.toa_swdn - flux_ds.toa_swup - flux_ds.toa_lwup

fig = plt.figure()
ax = plt.axes(projection=geo_p)

img = ax.imshow(toa_flux[::-1,::-1][51:3660, 51:3667], extent=geo_extent, vmin=-1000, vmax=1000, cmap="RdBu_r")
ax.coastlines(alpha=0.5)
plt.colorbar(img, label="ToA flux balance [Wm$^{-2}$]")
title = plt.title("Derived ToA flux balance")
```

We can clearly see that during the daytime clouds cause a large reduction in the ToA heating. By subtracting the clear sky flux from the cloudy sky flux we can show the cloud radiative effect (CRE) quantitatively (note that there is no change in the SW downwelling flux)

During the daytime, CRE is mostly negative (cooling) due to the reflectance of incoming sunlight. During the night it is positive (warming) as cold clouds radiate less heat than the surface below them.

```python
toa_cre = - (flux_ds.toa_swup - flux_ds.toa_swup_clr) - (flux_ds.toa_lwup - flux_ds.toa_lwup_clr)

fig = plt.figure()
ax = plt.axes(projection=geo_p)

img = ax.imshow(toa_cre[::-1,::-1][51:3660, 51:3667], extent=geo_extent, vmin=-1000, vmax=1000, cmap="RdBu_r")
ax.coastlines(alpha=0.5)
plt.colorbar(img, label="ToA CRE [Wm$^{-2}$]")
title = plt.title("Derived ToA Cloud Radiative Effect")
```

```python
# First get daytime only retrievals
wh_day = cloud_ds.illum == 1

# And only over the oceans
wh_sea = cloud_ds.lsflag == 0
```

```python
# Then use a KDTree to find the nearest neighbouring SEVIRI pixel for each emission track location
from scipy.spatial import KDTree

wh_seviri_finite = np.isfinite(cloud_ds.lon.data.ravel())
seviri_lonlat_tree = KDTree(np.stack([cloud_ds.lon.data.ravel()[wh_seviri_finite], cloud_ds.lat.data.ravel()[wh_seviri_finite]], -1))

wh_emissions_finite = np.isfinite(emissions_advected.longitude.to_numpy())
dists, nearest_neighbours = seviri_lonlat_tree.query(np.stack([
    emissions_advected.longitude.to_numpy()[wh_emissions_finite], 
    emissions_advected.latitude.to_numpy()[wh_emissions_finite]
], -1))

# Filter distances that are too large
dist_filter = dists < 0.1
nearest_neighbours = nearest_neighbours[dist_filter]

wh_overlap = wh_seviri_finite.copy()
wh_overlap[wh_overlap] = np.isin(np.arange(np.sum(wh_seviri_finite), dtype=int), nearest_neighbours)
wh_overlap = wh_overlap.reshape(cloud_ds.lon.shape)
```
